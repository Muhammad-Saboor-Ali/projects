```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import OneHotEncoder
from sklearn.preprocessing import LabelEncoder
```

```python
dataset = pd.read_csv("car_price.csv")
```

```python
dataset.drop(['Unnamed: 0'],axis=1,inplace=True)
```

```python
dataset.shape
```

```python
dataset.sample(10)
```

```python
dataset.isnull().sum()
```

```python
dataset['car_name'].value_counts()
```

```python
le = LabelEncoder()
```

```python
dataset['car_name'] = le.fit_transform(dataset['car_name'])
```

```python
def convert_lakh(value):
    if isinstance(value, str):
        value = value.lower().replace(',', '').strip()
        if 'lakh' in value:
            return float(value.replace('lakh', '')) * 100000
        elif 'crore' in value:
            return float(value.replace('crore', '')) * 10000000
    return float(value)

dataset['car_prices_in_rupee'] = dataset['car_prices_in_rupee'].apply(convert_lakh)
```

```python
dataset
```

```python
dataset['car_prices_in_rupee'] = dataset['car_prices_in_rupee'].astype(int)
```

```python
dataset['car_prices_in_rupee'].dtype
```

```python
dataset
```

```python
dataset['kms_driven'] = (dataset['kms_driven'].astype(str).str.replace(',', '').str.extract(r'(\d+)').astype(float))
```

```python
dataset['kms_driven'] = dataset['kms_driven'].astype(int)
```

```python
dataset['kms_driven']
```

```python
dataset.head()
```

```python
dataset['Seats'] = (
    dataset['Seats']
    .astype(str)
    .str.extract(r'(\d+)')
    .astype(int)
)
```

```python
dataset
```

```python
dataset['manufacture'].dtype
```

```python
one = OneHotEncoder(sparse_output=False,drop='first')
```

```python
encoded = one.fit_transform(dataset[['fuel_type','transmission','ownership']])
```

```python
encoded_df = pd.DataFrame(
    encoded,
    columns=one.get_feature_names_out(
        ['fuel_type', 'transmission', 'ownership']
    )
)
```

```python
dataset = dataset.drop(
    ['fuel_type', 'transmission', 'ownership'],
    axis=1
)

dataset = pd.concat([dataset, encoded_df], axis=1)
```

```python
dataset
```

```python
dataset['engine'] = pd.to_numeric(
    dataset['engine']
    .astype(str)
    .str.replace(',', '', regex=True)
    .str.extract(r'(\d+)')[0],
    errors='coerce'
)
```

```python
dataset['engine'].dtype
```

```python
dataset
```

```python
x = dataset.drop('car_prices_in_rupee',axis=1)
```

```python
y = dataset['car_prices_in_rupee']
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=87)
```

```python
standard = StandardScaler()
```

```python
x_train[["kms_driven","manufacture","engine"]] = standard.fit_transform(x_train[["kms_driven","manufacture","engine"]])
```

```python
x_test[["kms_driven","manufacture","engine"]] = standard.transform(x_test[["kms_driven","manufacture","engine"]])
```

```python
from sklearn.tree import DecisionTreeRegressor
```

```python
from sklearn.metrics import r2_score
```

```python
dc = DecisionTreeRegressor()
```

```python
dc.fit(x_train,y_train)
```

```python
y_pred = dc.predict(x_test)
```

```python
dc.score(x_train,y_train)*100
```

```python
r2_score(y_test,y_pred)*100
```

```python
for i in range(1,100):
    x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=i)
    dc1 = DecisionTreeRegressor()
    dc1.fit(x_train,y_train)
    y_pred = dc1.predict(x_test)
    print(dc1.score(x_train,y_train)*100,r2_score(y_test,y_pred)*100,i)
```
